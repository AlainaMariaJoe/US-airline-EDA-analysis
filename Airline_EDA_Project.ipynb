#### DATA CLEANING


```python
# Sum of missing values
df.isnull().sum()
```

```python
#drop rows where arr_flights is null( no flights)
df = df.dropna(subset=['arr_flights']).copy()
df.shape
```

```python
df['arr_del15'] = df['arr_del15'].astype(int)
df['arr_flights'] = df['arr_flights'].round().astype(int)
```

```python
# check negative values
count_cols = [
'carrier_ct','weather_ct','nas_ct',
'security_ct','late_aircraft_ct'
]

for col in count_cols:
    print(col, (df[col]< 0).sum())
```

```python
# fix negative values
df.loc[df['nas_ct'] < 0, 'nas_ct'] = 0
```

```python
# Rename carrier names(ensure consistent name)
import re
def clean_airline(name):
    name = re.sub(r'\s+(Inc\.?|Corporation|Co\.?|LLC|Network)$', '', name)
    return name.strip()

df.loc[:, 'carrier_clean'] = df['carrier_name'].apply(clean_airline)
df['carrier_clean'].unique()
```

```python
df.groupby('carrier')['carrier_clean'].unique()
```

```python
df.loc[:,'carrier_clean'] = df['carrier_clean'].replace({
    'Pinnacle Airlines': 'Endeavor Air',
    'Atlantic Southeast Airlines': 'ExpressJet Airlines',
    'American Eagle Airlines': 'Envoy Air',
    'Atlantic Coast Airlines': 'Independence Air',
    'Spirit Air Lines': 'Spirit Airlines',
    'Commutair Aka Champlain Enterprises,': 'CommuteAir',
    'CommuteAir LLC dba CommuteAir': 'CommuteAir',
    'GoJet Airlines LLC d/b/a United Express': 'GoJet Airlines',
    'ATA Airlines d/b/a ATA': 'ATA Airlines'
})
```

```python
df.groupby('carrier')['carrier_clean'].unique()
```

```python
df['carrier_clean'].nunique()
```

```python
# reduced the number of unique airline names from 52 to 37 (preventing duplicates)
```

### Airline Analysis

```python
# Top 5 airlines by no of flights
top_airlines = (
    df.groupby('carrier_clean')['arr_flights']
    .sum()
    .sort_values(ascending=False)
)

top_airlines.head()
```

```python
top_airlines.head().plot(kind='bar', color = "yellow")
plt.title('Top 5 Airlines by Number of Flights')
plt.ylabel('Number of Flights')
plt.xlabel('Airline')
plt.ticklabel_format(style='plain', axis='y')

plt.xticks(rotation=45)
for i, v in enumerate(top_airlines.head()):
    plt.text(i, v +v*0.02, str(v), ha='center', fontweight='bold')
plt.show()
```

```python
#Top 5 airlines by Most delay

delays_by_airline = (
    df.groupby('carrier_clean')['arr_del15']
    .sum()
    .sort_values(ascending=False)
)

delays_by_airline.head()
```

```python
delays_by_airline.head().plot(kind='bar', color = "green")
plt.title('Top 5 Airlines by Most delay')
plt.ylabel('Delay in min')
plt.xlabel('Airline')
plt.xticks(rotation=45)
plt.xticks(rotation=45)
for i, v in enumerate(delays_by_airline.head()):
    plt.text(i, v +v*0.02, str(v), ha='center', fontweight='bold')
plt.show()
```

```python
# Calculate delay percentage by Airline

df['delay_rate'] = np.where(
    df['arr_flights'] == 0,
    0,
    df['arr_del15'] / df['arr_flights']
)
df['delay_percentage'] = np.round(df['delay_rate'] * 100, 2)
```

```python
# Top 5 airlines delays passengers the most

most_delay_airline = (
    df.groupby('carrier_clean')['delay_percentage']
    .mean()
    .sort_values(ascending=False)
)

most_delay_airline.head(10)
```

```python

most_delay_airline_top10 = most_delay_airline.head(10).reset_index()

plt.figure(figsize=(10,6))
ax = sns.barplot(
    data=most_delay_airline_top10,
    y='carrier_clean',
    x='delay_percentage',
    hue = 'delay_percentage',
    palette='viridis'
)

plt.title("Top Airlines with Highest Delay Rate")
plt.xlabel("Average Delay %")
plt.ylabel("Airline")
for i, v in enumerate(most_delay_airline_top10['delay_percentage']):
    ax.text(v + 0.5, i, f"{v:.2f}%", fontweight='bold')

plt.show()
```

```python
# Top 5 airlines delays passengers the least

least_delay_airlines = (
    df.groupby('carrier_clean')['delay_percentage']
    .mean()
    .sort_values(ascending=True)
)

least_delay_airlines.head()
```

```python
least_delay = least_delay_airlines.head(5).reset_index()
plt.figure(figsize=(8,8))
plt.pie(
    least_delay['delay_percentage'],
    labels=least_delay['carrier_clean'],
    autopct='%1.1f%%',
    startangle=140
)
plt.title("Top 5 Airlines with Lowest Delay %")
plt.legend(
    least_delay['carrier_clean'],
    title="Airlines",
    loc="upper right",
    bbox_to_anchor=(1, 0, 0.5, 1)
)

plt.show()
```

### Delay Cause Analysis

```python
delay_causes = df[
    ['carrier_ct','weather_ct','nas_ct','security_ct','late_aircraft_ct']
].sum()

delay_causes = delay_causes.sort_values(ascending=False)
delay_causes
```

```python
delay_causes.plot(kind='bar', color = "Red")

plt.title("Flight Delays by Cause")
plt.xlabel("Delay Cause")
plt.ylabel("Total Delays")

plt.xticks(rotation=45)

plt.ticklabel_format(style='plain', axis='y')
for i, v in enumerate(delay_causes):
    plt.text(i, v + v*0.02, str(int(v)), ha='center', fontweight='bold')

plt.show()
```

### Airpot Analysis

```python
# Top 5 airports by no of flights
top_airports = (
    df.groupby('airport')['arr_flights']
    .sum()
    .sort_values(ascending=False)
)

top_airports.head()
```

```python
top_airports.head().plot(kind='bar', color = "orange")

plt.title("Top 5 Airports by Number of Flights")
plt.xlabel("Airport")
plt.ylabel("Total Flights")

plt.xticks(rotation=45)
plt.ticklabel_format(style='plain', axis='y')
for i, v in enumerate(top_airports.head()):
    plt.text(i, v + v*0.02, str(int(v)), ha='center', fontweight='bold')

plt.show()
```

```python
# airpot delay rate
airport_delay = (
    df.groupby('airport')['delay_rate']
    .mean()
    .sort_values(ascending=False)
)
airport_delay.head()
```

```python

top_delay_airports = airport_delay.head(10).reset_index()

sns.barplot(
    data=top_delay_airports,
    y='airport',
    x='delay_rate',
    hue='delay_rate',
    palette='viridis',
    dodge=False,
    legend=False
)

plt.title("Airports with Highest Delay Rate")
plt.xlabel("Average Delay Rate")
plt.ylabel("Airport")

for i, v in enumerate(top_delay_airports['delay_rate']):
    plt.text(v + 0.1, i, f"{v:.2f}", va='center', fontweight='bold')

plt.show()
```

```python
# Top 5 airports by lowest delay
least_delay_airports = (
    df.groupby('airport')['delay_rate']
    .mean()
    .sort_values()
)
least_delay_airports.head(10)
```

```python
least_delay_airports.head(10).plot(kind='bar', color = "purple")

plt.title("Top 10 Airports with Lowest Delay Rate")
plt.xlabel("Airport")
plt.ylabel("Average Delay Rate")

plt.show()
```

```python
#Correlation between delay causes

delay_cols = df[
    ['carrier_ct','weather_ct','nas_ct','security_ct','late_aircraft_ct']
]
corr_matrix = delay_cols.corr()
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Between Delay Causes")

plt.show()
```

```python
# Carrier delays and late aircraft delays have a very strong relationship (0.85)
```

```python
sns.boxplot(x=df['delay_rate'])

plt.title("Distribution of Flight Delay Rate")

plt.xlabel("Delay Rate")

plt.show()
```

###  Key Insights from EDA

```python
# 1. Airline Performance
# Southwest, Delta, and American Airlines are the busiest carriers.

# 2. Delay Causes
# Late aircraft delays are the most common reason, while security delays are the least.
# Carrier delays and late aircraft delays are highly correlated (~0.85)

# 3. Airport Performance
# ATL, ORD, and DFW handle the highest traffic.
# Some medium-sized airports have high delay rates, whereas a few high-traffic airports maintain low delays.

# 4. Distribution & Outliers
# Most flights experience moderate delays, but a few outliers indicate rare extreme cases.

# 5. Overall Insight
# Improving one factor (like carrier operations) can positively affect others (like late aircraft and NAS delays).
```
